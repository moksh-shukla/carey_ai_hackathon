```python
def calculate_win_loss_rates(df, groupby_column):
    closed_df = df[df['Opportunity_Stage'].isin(['Closed Won', 'Closed Lost'])]

    grouped = closed_df.groupby([groupby_column, 'Opportunity_Stage']).size().unstack()

    grouped = grouped.fillna(0)

    grouped['Total'] = grouped.sum(axis=1)
    grouped['Win_Rate'] = grouped['Closed Won'] / grouped['Total']
    grouped['Loss_Rate'] = grouped['Closed Lost'] / grouped['Total']

    return grouped[['Win_Rate', 'Loss_Rate']]


win_loss_rates_by_type = calculate_win_loss_rates(df, 'Opportunity_Type')
print(win_loss_rates_by_type)

print('\n')

win_loss_by_product = calculate_win_loss_rates(df, 'Opportunity_Product')
print(win_loss_by_product)

print('\n')

win_loss_by_emr = calculate_win_loss_rates(df, 'Account_Electronic_Medical_Record_System')
print(win_loss_by_emr)

print('\n')

win_loss_by_hospital_type = calculate_win_loss_rates(df, 'Account_Hospital_Type')
print(win_loss_by_hospital_type)
```

```python
closed_won_accounts = df[df['Opportunity_Stage'] == 'Closed Won']['Opportunity_ID'].unique()

total_accounts = df['Opportunity_ID'].unique()

percentage_accounts_closed_won = len(closed_won_accounts) / len(total_accounts) * 100

print(f"Percentage of businesses with at least one 'Closed Won': {percentage_accounts_closed_won:.2f}%")
```

```python
df['Opportunity_Created_Date'] = pd.to_datetime(df['Opportunity_Created_Date'])
df['Opportunity_Close_Date'] = pd.to_datetime(df['Opportunity_Close_Date'])

df['Stage_Duration'] = (df['Opportunity_Close_Date'] - df['Opportunity_Created_Date']).dt.days

average_duration_by_stage = df.groupby('Opportunity_Stage')['Stage_Duration'].mean()

print('By days:')
print(average_duration_by_stage)
```

```python
df['Duration'] = (df['Opportunity_Close_Date'] - df['Opportunity_Created_Date']).dt.days

bins = pd.cut(df['Duration'], bins=[0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, max(df['Duration'])])

win_rate_by_duration = df.groupby([bins, 'Opportunity_Stage']).size().unstack(fill_value=0)

for stage in win_rate_by_duration.columns:
    total_in_stage = win_rate_by_duration[stage].sum()
    win_rate_by_duration[f'{stage}_Rate_Percentage'] = (win_rate_by_duration[stage] / total_in_stage) * 100

print(win_rate_by_duration.filter(like='_Rate_Percentage'))
```
